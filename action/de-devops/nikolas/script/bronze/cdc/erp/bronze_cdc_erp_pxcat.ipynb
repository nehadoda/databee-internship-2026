# Bronze Layer: ERP Product Category (Auto Loader)

**Pattern**: Auto Loader (Streaming Ingestion)  
**Trigger**: AvailableNow (Incremental Batch)

```python
from pyspark.sql.functions import current_timestamp

# Pointing to the directory, not the file
source_path = "/Volumes/workspace/bronze/raw_sources/source_erp/"
target_table = "workspace.bronze.erp_px_cat_g1v2_cdc"
checkpoint_path = "/Volumes/workspace/bronze/raw_sources/checkpoints/erp_pxcat"
schema_path = "/Volumes/workspace/bronze/raw_sources/checkpoints/erp_pxcat_schema"

query = (
    spark.readStream.format("cloudFiles")
    .option("cloudFiles.format", "csv")
    .option("cloudFiles.header", "true")
    .option("cloudFiles.inferColumnTypes", "true")
    .option("cloudFiles.schemaLocation", schema_path)
    .option("pathGlobFilter", "PX_CAT_G1V2.csv")
    .load(source_path)
    .withColumn("ingestion_timestamp", current_timestamp())
    .writeStream
    .format("delta")
    .outputMode("append")
    .option("checkpointLocation", checkpoint_path)
    .trigger(availableNow=True)
    .toTable(target_table)
)

query.awaitTermination()
```
