# Bronze Layer: Sales Details (Auto Loader)

**Pattern**: Auto Loader (Streaming Ingestion)  
**Trigger**: AvailableNow (Incremental Batch)

```python
from pyspark.sql.functions import current_timestamp

source_path = "/Volumes/workspace/bronze/raw_sources/source_crm/"
target_table = "workspace.bronze.crm_sales_details_cdc"
checkpoint_path = "/Volumes/workspace/bronze/raw_sources/checkpoints/crm_sales_details"
schema_path = "/Volumes/workspace/bronze/raw_sources/checkpoints/crm_sales_details_schema"

query = (
    spark.readStream.format("cloudFiles")
    .option("cloudFiles.format", "csv")
    .option("header", "true")
    .option("cloudFiles.inferColumnTypes", "true")
    .option("cloudFiles.schemaLocation", schema_path)
    .option("pathGlobFilter", "sales_details.csv")
    .load(source_path)
    .withColumn("ingestion_timestamp", current_timestamp())
    .drop("_rescued_data")
    .writeStream
    .format("delta")
    .outputMode("append")
    .option("checkpointLocation", checkpoint_path)
    .trigger(availableNow=True)
    .toTable(target_table)
)

query.awaitTermination()
```
