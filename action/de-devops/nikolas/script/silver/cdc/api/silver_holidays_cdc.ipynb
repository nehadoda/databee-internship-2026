# Silver Layer: Holidays API (Simple Overwrite)

**Pattern**: Overwrite  

This notebook cleans the holiday data and overwrites the silver table. Simple mode for Free Edition.

```python
from pyspark.sql.functions import col

print("Reading from Bronze...")
df = spark.table("workspace.bronze.holidays_api_cdc").select(
    col("country_code"),
    col("date").cast("date"),
    col("name").alias("holiday_name"),
    col("local_name"),
    col("is_global"),
    col("ingestion_timestamp")
)

print("Saving to Silver (Overwrite)...")
df.write.format("delta").mode("overwrite").saveAsTable("workspace.silver.holidays_cdc")
print("✓ Done.")
```
