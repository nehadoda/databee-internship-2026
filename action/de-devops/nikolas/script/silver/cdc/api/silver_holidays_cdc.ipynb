# Silver Layer: Holidays (Simple Overwrite)

**Pattern**: Overwrite (Full Load)

```python
from pyspark.sql.functions import col

bronze_table = "workspace.bronze.holidays_api_cdc"
silver_table = "workspace.silver.holidays_cdc"

print("Transforming Holidays to Silver...")
df = spark.table(bronze_table).select(
    col("date").cast("date"),
    col("local_name").alias("holiday_name"),
    col("country_code"),
    col("ingestion_timestamp")
)

df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(silver_table)
print("✓ Done.")
```
